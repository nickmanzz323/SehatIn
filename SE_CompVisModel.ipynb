```python
from ultralytics import YOLO
from IPython.display import display, Image
import os

# Define the paths to your two trained models
model_path_1 = "/content/50Epoch.pt"
model_path_2 = "/content/ModelSehatin100EPoch.pt" # Replace with the path to your second model

# Define the paths to your three test images
test_image_paths = [
    "/content/ini-resep-mie-goreng-dengan-bahan-sederhana-yang-menggugah-selera-halodoc.jpg",
    "/content/nasi-goreng-bumbu-rendang-sate-udang-foto-resep-utama.jpg",  # Replace with the path to your second test image
    "/content/440331098_18409575046065555_8043666969149579631_n.jpg"   # Replace with the path to your third test image
]

# Load the two trained models
model_1 = YOLO(model_path_1)
model_2 = YOLO(model_path_2)

# Iterate through each test image
for i, test_image_path in enumerate(test_image_paths):
    print(f"Processing Image {i+1}: {os.path.basename(test_image_path)}")

    # Run inference on the current test image with both models
    results_1 = model_1.predict(source=test_image_path)
    results_2 = model_2.predict(source=test_image_path)

    # Get the results for the single image
    result_1 = results_1[0]
    result_2 = results_2[0]

    # Define output paths for the result images
    output_image_path_1 = f"output_image_{i+1}_model_1.jpg"
    output_image_path_2 = f"output_image_{i+1}_model_2.jpg"

    # Save the result images with detections
    result_1.save(output_image_path_1)
    result_2.save(output_image_path_2)

    # Display the output images side-by-side
    print("Results from Model 1:")
    display(Image(filename=output_image_path_1))

    print("\nResults from Model 2:")
    display(Image(filename=output_image_path_2))

    print("-" * 50) # Separator for clarity
```
